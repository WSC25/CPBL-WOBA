```python
def calculate_base_state(row):
    state = 0
    if pd.notna(row['firstId']):
        state += 1
    if pd.notna(row['secondId']):
        state += 2
    if pd.notna(row['thirdId']):
        state += 4
    return state

# 应用函数创建新的 "壘包狀態" 列
non_track['壘包狀態'] = non_track.apply(calculate_base_state, axis=1)
```

```python
non_track
```

```python
def get_expected_runs(row):
    base_type = row["壘包狀態"]
    outs = row["outs"]
    #balls = row["balls"]
    #strikes = row["strikes"]
    
    matching_row = RE[(RE["壘包狀態"] == base_type) & 
                          (RE["outs"] == outs)] 
                          #(RE["Balls"] == balls) & 
                          #(RE["Strikes"] == strikes)]
    
    if not matching_row.empty:
        return matching_row.iloc[0]["RE"]
    else:
        return None

non_track["RE"] = non_track.apply(get_expected_runs, axis=1)

# 打印结果
non_track
```

```python
non_track['打者編號'] = non_track.groupby(['gameId', 'side', 'inning']).cumcount() + 1
non_track['打者編號'] = non_track.groupby(['gameId', 'side', 'inning']).cumcount() + 1
```

```python
non_track = non_track.dropna(subset=['result'])
```

```python
non_track['RE_change'] = non_track.groupby(['gameId','inning','side'])['RE'].diff().shift(-1)

last_bat_indexes = non_track.groupby(['gameId', 'inning', 'side'])['打者編號'].idxmax()

non_track.loc[last_bat_indexes, 'RE_change'] = -non_track['RE']
```

## 計算聯盟WOBA

```python
result_pivot = non_track.groupby('bName')['result'].value_counts().unstack(fill_value=0)
#result_pivot = result_counts.pivot_table(index=['bName','result'], columns='result', values='count', fill_value=0).reset_index()
result_pivot
```

```python
result_pivot['AB']=result_pivot['1B']+result_pivot['2B']+result_pivot['3B']+result_pivot['HR']+result_pivot['E']+result_pivot['FC']+result_pivot['FO']+result_pivot['GIDP']+result_pivot['GO']+result_pivot['SO']
result_pivot['PA']=result_pivot['AB']+result_pivot['HBP']+result_pivot['IBB']+result_pivot['SH']+result_pivot['SF']+result_pivot['uBB']
result_pivot['AVG']=(result_pivot['1B']+result_pivot['2B']+result_pivot['3B']+result_pivot['HR'])/result_pivot['AB']
result_pivot['OBP']=(result_pivot['1B']+result_pivot['2B']+result_pivot['3B']+result_pivot['HR']+result_pivot['HBP']+result_pivot['IBB']+result_pivot['uBB'])/result_pivot['PA']
result_pivot['SLG']=(result_pivot['1B']+result_pivot['2B']*2+result_pivot['3B']*3+result_pivot['HR']*4)/result_pivot['AB']
result_pivot['OPS']=result_pivot['SLG']+result_pivot['OBP']
result_pivot['ISO']=result_pivot['SLG']-result_pivot['AVG']
result_pivot
```

```python
league_woba = (
    summary['RE24']['1B'] * result_pivot['1B'].sum() +
    summary['RE24']['2B'] * result_pivot['2B'].sum() +
    summary['RE24']['3B'] * result_pivot['3B'].sum() +
    summary['RE24']['HR'] * result_pivot['HR'].sum() +
    summary['RE24']['uBB'] * result_pivot['uBB'].sum() +
    summary['RE24']['HBP'] * result_pivot['HBP'].sum()
) / (
    result_pivot['AB'].sum() + 
    result_pivot['uBB'].sum() - 
    result_pivot['IBB'].sum() + 
    result_pivot['SF'].sum() + 
    result_pivot['HBP'].sum()
)
```

```python
league_woba
```

```python
league_obp = (result_pivot['1B'].sum()+result_pivot['2B'].sum()+result_pivot['3B'].sum()+result_pivot['HR'].sum()+result_pivot['HBP'].sum()+result_pivot['IBB'].sum()+result_pivot['uBB'].sum())/result_pivot['PA'].sum()
woba_scale = league_obp/league_woba
print(f'聯盟上壘率: {league_obp}')
print(f'Woba_scale: {woba_scale}')
```

```python
result_pivot['woba']= ((
    woba_scale * summary['RE24']['1B'] * result_pivot['1B'] +
    woba_scale * summary['RE24']['2B'] * result_pivot['2B'] +
    woba_scale * summary['RE24']['3B'] * result_pivot['3B'] +
    woba_scale * summary['RE24']['HR'] * result_pivot['HR'] +
    woba_scale * summary['RE24']['uBB'] * result_pivot['uBB'] +
    woba_scale * summary['RE24']['HBP'] * result_pivot['HBP']
) / (
    result_pivot['AB'] + 
    result_pivot['uBB'] - 
    result_pivot['IBB'] + 
    result_pivot['SF']+ 
    result_pivot['HBP']
))
```
